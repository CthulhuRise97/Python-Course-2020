```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline

vuelos = sns.load_dataset('flights')
vuelos.head()
```

```python
vuelos = vuelos.pivot('month','year','passengers')
vuelos
```

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos, annot=True, fmt='d')
```

```python
val_central = vuelos.loc['May'][1956]
val_central
```

```python
sns.heatmap(vuelos,center=val_central,annot=True,fmt='d')
```

```python
sns.heatmap(vuelos,center=val_central,annot=True,fmt='d',cbar_kws={'orientation':'horizontal'})
```
