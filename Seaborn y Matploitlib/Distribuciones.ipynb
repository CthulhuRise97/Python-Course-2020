```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns

datos = np.random.randn(100)
datos
```

```python
sns.distplot(datos)
```

```python
sns.distplot(datos, color='green',rug=False,hist=False)
```

```python
arg_curva = {'color':'black','label':'curva'}
arg_histo = {'color':'grey','label':'histograma'}

sns.distplot(datos,bins=25,kde_kws=arg_curva,hist_kws=arg_histo)
```

```python
serie = pd.Series(datos)
serie
```

```python
sns.distplot(serie,bins=25,color='green')
```
