```python
#Ejercicio 1
#Filtrar datos en un dataframe.
#
#Crear una lista con 50 valores aleatorios enteros entre 0 y 100
#Convertir en un dataframe de 5x10
#Filtrar los datos para visualizar solo los valores mayores a 50
```

```python
import numpy as np
import pandas as pd

lista_valores = np.random.randint(0,100,50)
lista_valores.resize(5,10)

dataframe = pd.DataFrame(lista_valores)
dataframe
```

```python
dataframe[dataframe > 50]
```
